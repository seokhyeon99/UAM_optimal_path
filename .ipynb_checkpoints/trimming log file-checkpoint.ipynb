## Reading raw file

```python
file_location_and_name = '/Desktop/wide_Left_1 - 2021_05_07__1527__hedgehogs.csv' #raw log file name
```

```python
import pandas as pd
import csv
import os

userhome = os.path.expanduser('~')
csvfile = userhome + file_location_and_name
df = pd.read_csv(csvfile) #read raw log file as a dataframe using pandas
```

```python
df
```

## Processing log file

```python
df.drop(df.iloc[:,7:] , axis=1, inplace=True)
df.drop(df.iloc[:,1:4] , axis=1, inplace=True)
#Remove other values than Unix time and x, y, and z coordinates.
```

```python
df
```

```python
import datetime
time = []
x = []
y = []
z = []
for i in range(0,len(df)):
    #Convert Unix time to real time
    timestamp = datetime.datetime.fromtimestamp(df.iloc[i,0]/1000+4*3600)
    time.append(timestamp.strftime('%Y-%m-%d %H:%M:%S'))
    #Multiply each coordinate by 1000 and then get the quotient of each devided by 10
    x.append(int(df.iloc[i,1]*1000//10))
    y.append(int(df.iloc[i,2]*1000//10))
    z.append(int(df.iloc[i,3]*1000//10))

    
temp = {'time':  time,
          'x': x,
          'y' : y,
          'z' : z
         }

datas = pd.DataFrame(temp, columns= ['time','x','y','z'])
```

```python
datas
```

## Removing garbage values

```python
trash = []
#If the difference between the previous value and current value is greater than 100,the current value is removed.
#Then the previous value and the next value are compared again.
for i in range(1,len(datas)):
    for j in ('x', 'y', 'z'):
        if datas[j][i] > datas[j][i-1] + 100 or datas[j][i] < datas[j][i-1] - 100:
            datas[j][i] = datas[j][i-1]
            trash.append(i)

datas = datas.drop(datas.index[trash])
```

```python
print('deleted row :',trash)#Show which rows are deleted
```

```python
#Write the processed data as a csv file
datas.to_csv(userhome+"/Desktop/wide_leftside_obstacle_data_1.csv", header=False, index=False)
```
