```python
file_location_and_name = '/Desktop/wide_left_3 - 2021_05_11__1612__hedgehogs.csv'
```

```python
import pandas as pd
import csv
import os

userhome = os.path.expanduser('~')
csvfile = userhome + file_location_and_name
df = pd.read_csv(csvfile)
```

```python
df
```

```python
df.drop(df.iloc[:,7:] , axis=1, inplace=True)
df.drop(df.iloc[:,1:4] , axis=1, inplace=True)
```

```python
df
```

```python
import datetime
time = []
x = []
y = []
z = []
for i in range(0,len(df)):
    timestamp = datetime.datetime.fromtimestamp(df.iloc[i,0]/1000+4*3600)
    time.append(timestamp.strftime('%Y-%m-%d %H:%M:%S'))
    x.append(int(df.iloc[i,1]*1000//10))
    y.append(int(df.iloc[i,2]*1000//10))
    z.append(int(df.iloc[i,3]*1000//10))

    
temp = {'time':  time,
          'x': x,
          'y' : y,
          'z' : z
         }

datas = pd.DataFrame(temp, columns= ['time','x','y','z'])
```

```python
datas
```

```python
trash = []
for i in range(1,len(datas)):
    for j in ('x', 'y', 'z'):
        if datas[j][i] > datas[j][i-1] + 100 or datas[j][i] < datas[j][i-1] - 100:
            datas[j][i] = datas[j][i-1]
            trash.append(i)

datas = datas.drop(datas.index[trash])
```

```python
print('deleted row :',trash)
```
